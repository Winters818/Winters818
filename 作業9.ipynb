```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from fredapi import Fred
```

```python
fred = Fred(api_key='36999b409383c231bb15a44d6384bdb4')
```

```python
mich_data = fred.get_series('MICH')
```

```python
unrate_data = fred.get_series('UNRATE')
```

```python
mich_data = mich_data.dropna().rename("MICH")
```

```python
unrate_data = unrate_data.dropna().rename("UNRATE")
```

```python
data = pd.concat([mich_data, unrate_data], axis=1)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['MICH'], label='預期通貨膨脹率 (MICH)', color='blue')
plt.plot(data.index, data['UNRATE'], label='失業率 (UNRATE)', color='red')
plt.title('預期通貨膨脹率與失業率')
plt.xlabel('時間')
plt.ylabel('百分比')
plt.legend()
plt.grid(True)
plt.show()
```
